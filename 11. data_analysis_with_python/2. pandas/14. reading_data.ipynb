### Reading Data From Different Sources


```python
#converting json into string
import pandas as pd
from io import StringIO
Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'
df=pd.read_json(StringIO(Data))
```

```python
df
```

```python
#converting df into json again
#by default orient will be based on index value
df.to_json()
```

```python
#index is default one
df.to_json(orient = 'index')
```

```python
#changing orient value into records
df.to_json(orient='records')
```

```python
# reading wikipedia page
import requests
from io import StringIO
import pandas as pd

url1 = "https://en.wikipedia.org/wiki/Mobile_country_code"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

# Get the page content with proper headers
response = requests.get(url1, headers=headers)
# Wrap the HTML content in StringIO
html_content = StringIO(response.text)
tables = pd.read_html(html_content, match='Country', header=0)

# Display the first matching table
tables[0]
```
