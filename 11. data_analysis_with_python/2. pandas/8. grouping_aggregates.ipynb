### what % of people from each country know python


```python
country_respondent= df['Country'].value_counts()
print(f"People who responds to the survery {country_respondent}")
```

```python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
print(f'Country uses python {country_uses_python}')
```

```python
python_df= pd.concat([country_respondent, country_uses_python], axis = 'columns', sort = False)
```

```python
python_df
```

```python
python_df.rename(columns={'count': 'number_of_respondent', 'LanguageWorkedWith': 'num_knows_python'}, inplace = True)
```

```python
python_df
```

```python
python_df['PctKnowsPython']= (python_df['num_knows_python']/python_df['number_of_respondent']) *100
```

```python
python_df
```

```python
#sort the countris with largest number of %
python_df.sort_values(by='PctKnowsPython', ascending=True, inplace = True)
```
