```python
people = {
    "first": ["rhs", 'Jane', 'John'], 
    "last": ["2", 'Doe', 'Doe'], 
    "email": ["rhs0@hotmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
#concatenate 2 rows toegther
df['first'] + " "+ df['last']
```

```python
#adding a column in dataframe
df['full_name']= df['first']+ ' '+df['last']
```

```python
df
```

```python
#dropping a column
df.drop(columns=['first', 'last'])
```

```python
df
```

```python
#we have to use inplace to make it permanent
df.drop(columns=['first', 'last'], inplace = True)
```

```python
df
```

```python
#we can use str method to split the full_name
df['full_name'].str.split(" ")
```

```python
#using expand argument to expand the name
df['full_name'].str.split(" ", expand = True)
```

```python
#now we add the 2 diff column to add into dataframe
#we have to put double bracket otherwise it will think like one column
df[['first', 'last']]= df['full_name'].str.split(" ", expand= True)
```

```python
df
```

```python
#using concat method to add a new 
r = pd.DataFrame([{'first': 'Tony'}])
df=pd.concat([df, r], ignore_index=True)
#here tony is added but others 2 rows are NaN value
```

```python
df
```

```python

people1 = {
    'first': ['Tony', 'Steve'], 
    'last': ['Stark', 'Rogers'], 
    'email': ['IronMan@avenge.com', 'Cap@avenge.com']
}
df2 = pd.DataFrame(people1)
```

```python
df2
```

```python
df = pd.concat([df, df2], ignore_index= True)
```

```python
df
```

```python
#dropping by index 
df.drop(index=3, inplace=True)
```

```python
df
```

```python
#we can also drop by argument and function
#also we will have to pass that into the df
filter1 = df['last']== 'Doe'
df= df.drop(index= df[filter1].index)
```
